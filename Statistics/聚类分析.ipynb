# 相似性的度量

考虑一组样本$X_{(n)}$,看作n个p纬欧式空间的n个点，他们之间的相似性可以用距离进行度量。
其中假设 $E(X_{(n)}) = \mu$, $var(X_{(n)})=\Sigma$.

## 数值变量相似性的度量

### 距离

- distance generated by L^p norm: 
- 
- Minkowski distance
$$
d(x_{(i)},x_{(j)}) =  (x_{(i)}-x_{(j)})^\prime \Sigma^{-1}(x_{(i)}-x_{(j)})
$$

### 相似性

$$
c(x,y) = \cos{\theta} = \frac{\inner{x}{y}}{\Vert x\Vert \Vert y\Vert}
$$

## 类别变量相似性的度量

$$
    匹配比例 = \frac{两个点拥有共同属性个数}{总属性个数}
$$

```python
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
plt.scatter(x, y)
plt.show()
```

## scipy

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
?linkage
```

```python
data = list(zip(x, y))
linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python
linkage_data
```

```python
?AgglomerativeClustering
```

## sklearn

- "euclidean", "l1", "l2", "manhattan", "cosine", or "precomputed".
- 'ward', 'complete', 'average', 'single'

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='l2', linkage='single')
labels = hierarchical_cluster.fit_predict(data)
```

```python
plt.scatter(x, y, c=labels)
```
