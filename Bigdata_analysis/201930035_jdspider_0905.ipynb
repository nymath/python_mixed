```python
import requests
from bs4 import BeautifulSoup 
import random
import json
#https://search.jd.com/Search?keyword=python&wq=python&page=3&s=56&click=0
#https://search.jd.com/Search?keyword=python&enc=utf-8&wq=python&pvid=63f3d08d6e8b4ffb941f4dc47e3be120
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36"}
lst_books = []
for pp in range(1,3,2):
    
    url = 'https://search.jd.com/Search?keyword={}&page={}'.format('python',pp)
    print(url)
    
    res = requests.get(url, headers=header)# requests.post(url, headers=Headers)
    res.encoding = res.apparent_encoding
    html = res.text
    
    #DOM
    soup = BeautifulSoup(html,'html5lib')

    lst_li = soup.find('div',id="J_goodsList").find_all('li')
    
    for idx, li in enumerate(lst_li):
        print(idx)
        book = []
        try:
            book_id = li['data-sku']
            book_price = li.find('div',class_='p-price').strong.i.string.strip()
            book_name =  "".join(li.find('div',class_='p-name').a.em.strings)
            book.append(book_id)
            book.append(book_price)
            book.append(book_name)            
        except:
            print("Error in", li )
            
        try:
            header['referer']='https://item.jd.com/{}.html#comment'.format(book_id)
            url_cm = 'https://club.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98&productId={}&score=0&sortType=5&page=1&pageSize=10&isShadowSku=0&rid=0&fold=1'.format(book_id)
            res = requests.get(url_cm,headers=header)
            jd = json.loads(res.text.lstrip('fetchJSON_comment98(').rstrip(');'))
            data_list = jd['comments']
            comments = " ".join([ data['content'] for data in data_list])
            book.append(comments)
        except:
            print("Error in Comments", book_id)
        lst_books.append(book)

print(lst_books)               
```
