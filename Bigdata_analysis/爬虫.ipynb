### Html分析的结构

使用soup.find(_all)去定位<b>,可以加上限定条件


```python
html = '''
<a>
<b id='True' class='1' price='29.8'>"数学分析"</b>
<b id='True' class='1' price='19.8'>"高等代数"</b>
</a>'''
soup = BeautifulSoup([html][0],"html5lib") #Dom数 document object model
soup.find('b')
#print(soup.prettify())
print(soup.find('b').string) #获取内容
#print(soup.find('b')['price']) 获取标签
```

```python
#找多个b
lst_b = soup.find_all('b')
for b in lst_b:
    print(b.string) #提取文本
    print(b['price']) #提取属性
```

```python
#找满足条件的b
#找多个b
lst_b = soup.find_all('b',id='True')
for b in lst_b:
    print(b.string)
    print(b['price'])
```

```python
#定位某个属性的值，如果与python的关键词相同，则需加入_
lst_b = soup.find_all('b',class_='1')
for b in lst_b:
    print(b.string)
    print(b['price'])
```

### 掉包爬取，定位

\begin{enumerate}
\item 定位，标签夹住的文本 <b> Python </b>
\item 标签某属性的东西 <b> price='29.8'Python </b>
\end{enumerate}


```python
url0 = 'http://search.dangdang.com/?key=Mathematicas&act=input'
n = 10
html = []
for i in range(1,n+1):
    url = url0 + str(i)
    print(url)
    html.append([requests.get(url).text])
    #print(html)
```

```python
#找到所有书的标签
for ht in html:
    soup_html = BeautifulSoup(ht[0],'html5lib')
    #因为我们不知道a怎么去定位，所以选择去找div，用关键词id去定位div
    lst_li = soup_html.find('div',id="search_nature_rg").find_all('li')
    for li in lst_li:
        print(li.find('a',dd_name='单品图片')['title'])
        print(li.find('span',class_='search_now_price').string.lstrip('¥'))
        #print(li.find('a',dd_name='单品作者')['title'])
        print(li.find('p',class_="search_book_author").find('a')['title'])
        print(li.find('a href',dd_name="单品出版社"))
```
