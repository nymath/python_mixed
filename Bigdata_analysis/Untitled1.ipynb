```python
import requests #urllib urllib2
from bs4 import BeautifulSoup
import html5lib
```

```python
for i in range(1,100):
    url= 'https://www.runoob.com/python/python-exercise-example{}.html'.format(i)
    html = requests.get(url).text
    soup_html = BeautifulSoup(html,'html5lib')
    exam = soup_html.find('div',class_="article-intro").find_all('span')
    lstss = list(map(lambda x:x.string,exam))
    print(soup_html.find('div',class_="article-intro").find_all('p')[1])
    print(''.join(lstss))
    print('*******************************************************************************')
```

```python
# 顺序字符串幂集
def opowerset(s):
    power_list = []
    for i in range(1,len(s)+1):
        for j in range(0,len(s)-1):
            power_list.append(s[j:j+i])
    return list(set(power_list))
opowerset('1234')

# 顺序集合幂集
def opowerset(s):
    power_list = []
    for i in range(1,len(s)+1):
        for j in range(0,len(s)-1):
            power_list.append(s[j:j+i])
    return list(set(power_list))
```

```python
def febonaci(n):
    if n == 1 or n == 2:
         return n
    else:
         return febonaci(n-1)+febonaci(n-2)
febonaci(3)
lst = [febonaci(n+2)/febonaci(n+1) for n in range(20)]
lst
```

```python
s1 = 'aaabbbbccc'
s2 = 'aaabbbbccc'
lst1 = list(s1)
lst2 = list(s2)
set1 = list(set(list(s1)))
set2 = list(set(list(s2)))
val1 = [lst1.count(i) for i in set1]
val2 = [lst2.count(i) for i in set2]
```

```python
comb1 = {}
for key,value in zip(set1,val1):
    comb1[key] = value
comb1
comb2 = {}
for key,value in zip(set2,val2):
    comb2[key] = value
comb2
comb1 == comb2
```
