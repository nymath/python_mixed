## Normal

```python
X = stats.norm(1,0.5)
```

```python
[X.moment(i) for i in range(10)]
```

```python
X_sample = X.rvs(100)
```

```python
stats.norm.fit(X_sample)
```

```python
x = np.linspace(-5,5,1000)
y1 = X.pdf(x)
y2 = X.cdf(x)
y3 = X.ppf(x)
fig, ax = plt.subplots(1,3,figsize=(16,4))
ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
```

```python
X.interval(0.999)
```

```python
?stats.multivariate_normal
```

```python
mu = np.array([1,1])
cov = np.array([[1,1.2],
                [1.2,2]])
```

```python
mn = stats.multivariate_normal(mu,cov)
```

```python
mn.cdf([1,1]) # 输入一个list
```

```python
mn.cov_info
```

```python
mn.entropy()
```

```python
mn.rvs(1000)
```
