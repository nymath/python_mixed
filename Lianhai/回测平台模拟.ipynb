```python
key = ["OpenPrice", "HighPrice", "LowPrice", "ClosePrice"]
ldf = {}
for k in key:
    ldf[k] = np.log(dtf[k])
```

## 法1

```python
# 假设为0
np.sqrt(np.nanmean((ldf['HighPrice']-ldf['LowPrice'])**2,axis=0))
```

## 法二

```python
0.5*np.nanvar(ldf['HighPrice']-ldf['LowPrice'],axis=0)-(2*np.log(2)-1)*np.nanvar(ldf['ClosePrice']-ldf['OpenPrice'],axis=0)
```

```python
np.nanmean((ldf['HighPrice']-ldf['LowPrice'])**2,axis=0)-(2*np.log(2)-1)*np.nanmean((ldf['ClosePrice']-ldf['OpenPrice'])**2,axis=0)

```

```python
np.sum(np.sum(cond!=1,0))
```

```python
from scipy.ndimage import shift
```

```python
ldf['ClosePrice_lag_1'] = np.apply_along_axis(lambda x:shift(x,1,cval=np.nan),0,ldf['ClosePrice'])
```

```python
np.nanmean(ldf['ClosePrice_lag_1'],0)
```

## 法三

```python
dfddd = np.sqrt(np.nanmean((ldf['HighPrice']-ldf['LowPrice'])**2,axis=0)- \
(2*np.log(2)-1)*np.nanmean((ldf['ClosePrice']-ldf['OpenPrice'])**2,axis=0)+ \
np.nanmean((ldf['OpenPrice']-ldf['ClosePrice_lag_1'])**2,axis=0))
```

```python
0.5*np.nanvar(ldf['HighPrice']-ldf['LowPrice'],axis=0)-\
(2*np.log(2)-1)*np.nanvar(ldf['ClosePrice']-ldf['OpenPrice'],axis=0)+\
np.nanvar(ldf['OpenPrice']-ldf['ClosePrice_lag_1'])
```

```python
cond1 = (ldf['HighPrice']-ldf['LowPrice'] > 0)*1
cond2 = (ldf['ClosePrice']-ldf['OpenPrice'] > 0)*1
cond3 = (ldf['OpenPrice']-ldf['ClosePrice_lag_1'] > 0)*1
np.sqrt(np.nanmean(cond1*0.5*(ldf['HighPrice']-ldf['LowPrice'])**2,axis=0)- \
    cond2*(2*np.log(2)-1)*np.nanmean((ldf['ClosePrice']-ldf['OpenPrice'])**2,axis=0)+ \
    cond3*np.nanmean((ldf['OpenPrice']-ldf['ClosePrice_lag_1'])**2,axis=0))
```
