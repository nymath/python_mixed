```python
import urllib.request
import re
import os


# open the url and read
def getHtml(url):
    page = urllib.request.urlopen(url)
    html = page.read()
    page.close()
    return html


def getFile(url):
    file_name = url.split('/')[-1]
    try:
        u = urllib.request.urlopen(url)
    except urllib.error.HTTPError:
        # 碰到了匹配但不存在的文件时，提示并返回
        print(url, "url file not found")
        return
    block_sz = 8192
    with open(file_name, 'wb') as f:
        while True:
            buffer = u.read(block_sz)
            if buffer:
                f.write(buffer)
            else:
                break
    print("Sucessful to download" + " " + file_name)


root_url = 'https://myweb.uiowa.edu/pbreheny/uk/teaching/760-s13/notes/'  


for j in range(1,13):
    for i in range(1,32):      # 任意给出控制号的范围，也可以根据自己单位的队伍号建立一个list
        url = root_url + str(j) + '-'+str(i) + '.pdf'  # 形成完整的下载地址
        getFile(url)
```
